```python
class TreeNode:
      def __init__(self, value):
            self.value = value
            self.left = None
            self.right = None
```

```python
node = TreeNode(10)
node.left = TreeNode(6)
node.right = TreeNode(15)
node.left.left = TreeNode(3)
node.left.right = TreeNode(8)
node.left.left.left = TreeNode(1)
node.right.right = TreeNode(20)
```

```python
# recursive
def dfs_preorder(node):
      if node is None:
            return
      
      print(node.value, end=" ")
      dfs_preorder(node.left)
      dfs_preorder(node.right)
```

```python
dfs_preorder(node)
```

```python
# recursive
def dfs_inorder(node):
      if node is None:
            return
      
      dfs_inorder(node.left)
      print(node.value, end=" ")
      dfs_inorder(node.right)
```

```python
dfs_inorder(node)
```

```python
# recursive
def dfs_postorder(node):
      if node is None:
            return
      
      dfs_postorder(node.left)
      dfs_postorder(node.right)
      print(node.value, end=" ")
```

```python
dfs_postorder(node)
```

```python
def dfs_preorder_iter(root):
      if root is None:
            return
      
      stack = [root]

      while stack:
            current = stack.pop()
            print(current.value, end=" ")

            if current.right:
                  stack.append(current.right)
            if current.left:
                  stack.append(current.left)
dfs_preorder_iter(node)
```

```python
def dfs_inorder_iter(root):
      stack = []
      current = root

      while current or stack:
            
            while current:
                  stack.append(current)
                  current = current.left

            current = stack.pop()
            print(current.value, end=" ")

            current = current.right
dfs_inorder_iter(node)
```
