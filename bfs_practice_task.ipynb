BFS Practice Task — Count Nodes Per Level

```python
from collections import deque
```

```python
class TreeNode:
      def __init__(self, value):
            self.value = value
            self.left = None
            self.right = None
```

Create this tree:  

            10
           /  \
          6    15
         / \     \
        3   8     20
       /
      1

```python
node = TreeNode(10)
node.left = TreeNode(6)
node.right = TreeNode(15)
node.left.left = TreeNode(3)
node.left.right = TreeNode(8)
node.left.left.left = TreeNode(1)
node.right.right = TreeNode(20)
```

```python
def count_nodes_per_level(root):
      if root is None:
            return
      
      queue = deque([root])

      while queue:
            level_length = len(queue)

            for _ in range(level_length):
                  current = queue.popleft()
                  print(f"Node value: {current.value}", end=" ")
                  
                  if current.left:
                        queue.append(current.left)
                  if current.right:
                        queue.append(current.right)
            print()
            print(f"Level length: {level_length}\n")
```

```python
count_nodes_per_level(node)
```
