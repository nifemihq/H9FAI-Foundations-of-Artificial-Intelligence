**Exercise 1**  
Ask the user to input for a number. Check whether the number is even or odd, print out an appropriate
message to the user.

```python
def even_or_odd(number):
      output = f"is even." if number % 2 == 0 else "is odd."
      print(f"{number} is {output}")

even_or_odd(int(input("Enter a number: ")))
```

**Exercise 2**  
Ask the user for a string and print out whether this string is a palindrome or not. A palindrome is a string
that reads the same forwards and backwards.

```python
def palindrome_checker(word):
      rev = word[::-1]
      output = f"a palindrome." if word.lower() == rev.lower() else "not a plaindrome."
      print(f"{word} is {output}")
      
palindrome_checker(input("Enter a word: "))
```

**Exercise 5**  
Write a program that will loop over the following list: a = [8, 9, 10, 11, 13, 81, 101, 100, 94]. If number is prime,
print its square. Use continue keyword.

```python
def check_prime(lst):
      for num in lst:
            if num < 2:
                  continue
            
            is_prime = True
            
            for i in range(2, int(num**0.5) + 1):
                  if num % i == 0:
                        is_prime = False
                        break
            
            if is_prime:
                  print(f"Prime number: {num}, Square: {num**2}")

a = [8, 9, 10, 11, 13, 81, 101, 100, 94]
check_prime(a)
```
