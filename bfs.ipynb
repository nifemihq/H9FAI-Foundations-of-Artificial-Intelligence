```python
from collections import deque
```

```python
class TreeNode:
      def __init__(self, value):
            self.value = value
            self.left = None
            self.right = None
```

```python
node = TreeNode(3)
node.left = TreeNode(7)
node.right = TreeNode(9)
node.left.left = TreeNode(5)
node.left.right = TreeNode(11)
node.right.left = TreeNode(14)
node.right.right = TreeNode(61)
```

```python
def bfs_print(node):
      if node is None:
            return
      
      queue = deque([node])

      while queue:
            current = queue.popleft()
            print(current.value, end=" ")

            if current.left:
                  queue.append(current.left)
            if current.right:
                  queue.append(current.right)
```

```python
bfs_print(node)
```
